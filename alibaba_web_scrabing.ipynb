```python
from selenium import webdriver
import pandas as pd
```

```python
scrapper =webdriver.Chrome()

scrapper.get("https://www.alibaba.com/trade/search?spm=a27aq.cp_44.4746171840.6.1a523ccfkPiijy&categoryId=201936701&SearchText=LED+%26+LCD+TVs&indexArea=product_en&fsb=y&productId=1601003345903")
```

```python
titles=[]
prices=[]
rates=[]
suppliers=[]
years=[]

print("========titles=========")
products_titles = scrapper.find_elements("xpath","//h2/a/span")
for item in products_titles:
    titles.append(item.text)
    print(item.text)

print("========prices=========")
products_prices = scrapper.find_elements("xpath","//div[@class='search-card-e-price-main']")
for item in products_prices:
    prices.append(item.text)
    print(item.text)

print("========rates=========")   
products_rates = scrapper.find_elements("xpath","//span[@class='search-card-e-review']/strong")
for item in products_rates:
    rates.append(item.text)
    print(item.text)

print("========suppliers=========")   
products_supplier= scrapper.find_elements("xpath","//span/span[2]")
for item in products_supplier:
    suppliers.append(item.text)
    print(item.text)
    
print("========years=========")   
products_years= scrapper.find_elements("xpath","//span[@class='margin-right-2']")
for item in products_years:
    years.append(item.text)
    print(item.text)
```

```python
print("========titles=========")
print(len(titles))

print("========prices=========")
print(len(prices))

print("========rates=========")   
print(len(rates))

print("========suppliers=========")   
print(len(suppliers))

print("========years=========")   
print(len(years))
```

```python
df = pd.DataFrame(titles,columns=['title'])
df['price']=prices
#df['rates']=rates  #not all products have rate
df['supplier']=suppliers
df['years']=years
df
```
