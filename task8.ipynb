```python
from selenium import webdriver
import pandas as pd
import time
```

# JUMIA SCRAPPING

```python
scrapper =webdriver.Chrome()
product_scrapper = webdriver.Chrome()
scrapper.get("https://www.jumia.com.eg")

titles=[]
prices=[]
rates=[]
links=[]

products = scrapper.find_elements("xpath","//article/a")
path_photos ="E:\Courses\DataExploration\Lab8\Jumia_photos\p"

for product in products:
    #print("url =",product.get_attribute('href'))
    product_scrapper.get(product.get_attribute('href'))
    time.sleep(2)
    titles.append(product_scrapper.find_elements("xpath","//div/h1")[0].text)
    prices.append(product_scrapper.find_elements("xpath","//div/span[@class='-b -ubpt -tal -fs24 -prxs']")[0].text)
    rates.append(product_scrapper.find_elements("xpath","//div[@class='stars _m _al']")[0].text)
    
    img = product_scrapper.find_elements("xpath","//*[@id='imgs']/a[1]/img")[0]
    links.append(img.get_attribute('src'))
    img.screenshot(path_photos+titles[-1] +'.png')
    
```

```python
print(len(titles))
print(len(prices))
print(len(rates))
print((len(links)))
```

```python
df = pd.DataFrame(titles,columns=['title'])
df['price']=prices
df['rates']=rates
df['img_link']=links
df.to_csv('task8-1.csv', index=False)
df
```

# AMAZON SCRAPPING

```python
scrapper =webdriver.Chrome()
product_scrapper = webdriver.Chrome()
scrapper.get("https://www.amazon.eg")
url = scrapper.find_element("xpath","//*[@id='nav-logo-sprites']").get_attribute('href')
#print("url = ",url)

scrapper.get(url)

titles=[]
prices=[]
rates=[]
links=[]

products = scrapper.find_elements("xpath","//div[@class='a-cardui dcl-product']/a[@class='a-link-normal dcl-product-link']")

path_photos ="E:\Courses\DataExploration\Lab8\Amazon_photos\A"

for product in products:
    #print("url =",product.get_attribute('href'))
    product_scrapper.get(product.get_attribute('href'))
    time.sleep(2)
    titles.append(product_scrapper.find_elements("xpath","//span[@id='productTitle']")[0].text)
    prices.append(product_scrapper.find_elements("xpath","//span[@class='a-price-whole']")[0].text)
    rates.append(product_scrapper.find_elements("xpath","//span[@class='a-icon-alt']")[0].text)
    
    img = product_scrapper.find_elements("xpath","//div[@class='imgTagWrapper']/img")[0]
    #print("img = ",img.get_attribute('src'))
    links.append(img.get_attribute('src'))
    img.screenshot(path_photos+titles[-1] +'.png')
    
```

```python
print(len(titles))
print(len(prices))
print(len(rates))
```

```python
df = pd.DataFrame(titles,columns=['title'])
df['price']=prices
df['rates']=rates
df['img_link']=links
df.to_csv('task8-2.csv', index=False)
df
```
